```python
def check_edu(x):
    if x in ['1st-4th', '5th-6th', 'Preschool', '9th', '7th-8th', '10th', '11th']:
        return 0
    else:
        return 1

df['atleast_high_school'] = df['education'].apply(lambda x: check_edu(x))
df.head()
```

```python
pd.crosstab(df['salary'], df['atleast_high_school'], normalize=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='atleast_high_school', hue='salary', data=df)
```

```python
pd.crosstab(df['salary'], df['atleast_high_school'])
```

```python
def check_marrried(x):
    if x in ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']:
        return 'Married'
    else:
        return 'Bachelors'
    
df['marital-status-new'] = df['marital-status'].apply(lambda x: check_marrried(x))
df.head()
```

```python
pd.crosstab(df['salary'], df['marital-status-new'], normalize=True)
```

```python
high_salary_df = df[df['salary'] > '50k']
proportional_by_marital_status = high_salary_df['marital-status-new'].value_counts(normalize=True)
proportional_by_marital_status
```
